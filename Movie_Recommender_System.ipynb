```python
# 1. Data
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
movies.head()
```

```python
credits.head()
```

```python
# 2. Data Preprocessing (EDA)
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

# Columns We need to build a model
1. Genres
2. id
3. keywords
4. title
5. overview
6. cast
7. crew

```python
movies = movies[['movie_id','title','genres','overview','cast','crew','keywords']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])

    return L
```

```python
convert('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else: 
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies['cast'][0]
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            L.append(i['name'])
            break;
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast']
```

```python
movies.head()
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
#new.head()
```

```python
new.head()
```

```python
new['tags'] = new['tags'].apply(lambda x: " ".join(x))
new.head()
```

```python
# 3. Model building
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
vector.shape
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
#!pip3 install nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
new['tags'] = new['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```

```python
similarity
```

```python
new[new['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend(movie):
    index = new[new['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1]) #enumerate to keep the indices while sorting at same place
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
        
```

```python
recommend('Avatar')
```
